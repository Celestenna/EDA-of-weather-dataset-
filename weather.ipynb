### EDA

```python
weather_df.head()
```

```python
top_10_count_of_weather= weather_df['Weather'].value_counts().reset_index().rename(columns={"index":"weather","Weather":"count"})[0:10]
top_10_count_of_weather
```

```python
sns.barplot(x="weather", y="count", data=top_10_count_of_weather)
plt.title("Count of Weather")
plt.xticks(rotation=90)
plt.show()
```

```python
top_10_count_of_Temp_C= weather_df['Temp_C'].value_counts().reset_index().rename(columns={"index":"Temp_C","Temp_C":"count"})[0:10]
top_10_count_of_Temp_C
```

```python
sns.histplot(x="Temp_C", y="count", data=top_10_count_of_Temp_C)
plt.title("Count of Temperature")
plt.xticks(rotation=90)
plt.show()
```
